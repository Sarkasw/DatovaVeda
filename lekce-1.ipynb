```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
data = pd.read_csv("hotel_bookings.csv")
```

```python
data.head()
```

```python
pobyty_do_deseti = data[data["stays_in_week_nights"] < 10]
pobyty_do_deseti.head()
```

```python
sns.countplot(data=pobyty_do_deseti, x="stays_in_week_nights")
```

```python
sns.countplot(data=pobyty_do_deseti, x="stays_in_week_nights", hue="is_canceled")
```

```python
data["total_lenght"] = data["stays_in_weekend_nights"] + data["stays_in_week_nights"]
data.head()
```

```python
pobyty_nenulove_do_deseti = data[(data["total_lenght"] > 0) & (data["total_lenght"] < 10)]
pobyty_nenulove_do_deseti.head()
```

```python
pobyty_nenulove_do_deseti.tail()
```

```python
sns.countplot(data=pobyty_nenulove_do_deseti, x="total_lenght")
```

```python
pobyty_country = pobyty_nenulove_do_deseti[["country", "adr", "is_canceled"]]
pobyty_country = pobyty_country[(pobyty_nenulove_do_deseti["adr"] < 300)]
pobyty_country["country"].value_counts()
```

```python
country_2 = pobyty_country[(pobyty_country["country"] == "GBR")]
sns.histplot(data=country_2, x="adr", hue="is_canceled")
```
